# single input node single output node linear layer 

```python
'''
[x] model defining
[X] auto gradient using optimizer
[x] constructing loss function
[x] constructing optimizer
[x] training loop
    [X] use model
    [x] calculate loss
    [X] use backward for auto gradient
    [x] use opimizer.step for auto parameter update
    [x] use opimizer.zero_grad
'''
import torch
import torch.nn as nn

# for linear modeling
# [[feature_vector],[feature_vector],[feature_vector],..]
#[sample1,sample2,sample3,..]

X = torch.tensor([[1],[2],[3],[4],[5]],dtype=torch.float32)
Y = torch.tensor([[2],[4],[6],[8],[10]],dtype=torch.float32)


#NOTE: weights will be automatically given by the nn.Linear


# number of input sample
# number of input feature or node
number_of_sample,number_of_input_feature=X.shape

number_of_input_node = number_of_input_feature

# number of output sample
# number of output feature or node

number_of_output_sample , number_of_output_feature = Y.shape

number_of_output_node = number_of_output_feature




# we need number of inputs to the linear layer and number of output from linear layer

# model = nn.Linear(in_features=number_of_input_node,out_features=number_of_output_node,bias=True)


# building linearRegression model 

# same as linear layer of nn.Module
class LinearRegression(nn.Module):

    def __init__(self,input_dim : int, out_dim : int):
        super(LinearRegression,self).__init__()
    
        self.linear_layer = nn.Linear(input_dim,out_dim)

    
    def forward(self,input):

        out_linear = self.linear_layer(input)

        return out_linear

model = LinearRegression(number_of_input_node,number_of_output_node)
# constructing loss and optimizer

# loss = MSE (Mean Square Error)
loss = nn.MSELoss()

# defining learning rate
learning_rate = 0.01
# optimizer = Stocastic Gradient Descent
optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)

X_test = torch.tensor([[10],[20],[30]],dtype=torch.float32)
Y_test_predict=model(X_test)
for (x,y) in zip(X_test,Y_test_predict):

    print(f"prediction before training  model({x[0]}) : {y[0]:.3f}")


iterations = 10000
# training
for epoch in range(iterations):

    
    # -----------------------forward pass----------------------
    # compute y_predict
    y_predict = model(X)

    # loss calculation
    l = loss(Y,y_predict)

    # ---------------------backward pass ------------------------
    # gradient calculation d(loss)/d(parameter)
    l.backward()

    # updating the parameters
    optimizer.step()

    #------------------- next pass grad zero --------------------
    # setting the gradients back to zero to avoid accumulation
    optimizer.zero_grad()

    if epoch % 10 == 0:

        [w,*rest] = model.parameters()


        print(f"------------------epoch {epoch+1}------------------------")
        print(f"loss: {l:0.3f} ______________ weight:{w[0][0]:0.3f}")


Y_test_predict=model(X_test)
for (x,y) in zip(X_test,Y_test_predict):

    print(f"prediction after training  model({x[0]}) : {y[0]:.3f}")


```
